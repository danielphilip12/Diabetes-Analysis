```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set the maximum number of rows to display to 100
pd.set_option('display.max_rows', None)

# Set the maximum number of columns to display to None (show all columns)
pd.set_option('display.max_columns', None)

sns.set_style('darkgrid')
```

```python
df = pd.read_csv('./data/diabetes_dataset.csv')
```

```python
sns.histplot(data=df, x='bmi', kde=True, hue='diabetes_stage')
```

```python
df['diabetes_stage'].value_counts()
```

```python
# Todo - remove gestational and type 1 diabetes, as the purpose of this analysis is for predicitng diabetes risk level as well asdiagnosis.
# Type 1 seems to be a sudden change that can occur at any time, ragardless of life style
# Gestational is pregenecy related, which also is not related to lifestyle. 
```

```python
remove = ['Type 1', 'Gestational']
```

```python
df_temp = df[~df['diabetes_stage'].isin(remove)]
```

```python
df_temp['diabetes_stage'].value_counts()
```

```python
df = df_temp
```

```python
del df_temp
```

```python
df.shape
```

```python
mean_bmi_by_stage = df.groupby('diabetes_stage')['bmi'].mean()
mean_bmi_by_stage
```

```python
sns.histplot(data=df, x='bmi', kde=True, hue='diabetes_stage', palette=['green', 'yellow', 'red'], hue_order=['No Diabetes', 'Pre-Diabetes', 'Type 2'])
plt.axvline(x = mean_bmi_by_stage['No Diabetes'], color='darkgreen')
plt.axvline(x = mean_bmi_by_stage['Pre-Diabetes'], color='darkgoldenrod')
plt.axvline(x = mean_bmi_by_stage['Type 2'], color='firebrick')
```

```python
plt.figure(figsize=(16, 16))
sns.heatmap(df.corr(numeric_only=True).round(2),
           vmin=-1,
           vmax=1,
           cmap='coolwarm',
           annot=True)
```

```python
sns.regplot(data=df, y='diabetes_risk_score', x='family_history_diabetes')
```
